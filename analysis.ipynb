```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 10)
y = x**2

plt.plot(x, y)
plt.title('Squares of Numbers')
plt.xlabel('X values')
plt.ylabel('X squared')

plt.savefig("output.png")   # SAVE FIRST
plt.show()
```
